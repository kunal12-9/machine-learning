```python
# supervised ML Model ---> classify 

# when we have target column as categorical data . 

# Logistic Regression Model -->
# Sigmod Activation function --> when we target data categorical and we have only 
# 2 - sub categories .

# sub categories ---> Probability ---> 0.1 ---
# if x> =0.5 :
#        return 1
# else :
#        return  0 

# performance Evaluation Metrics ---> Accuracy_score , precision , recall 
```

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
lb = LabelEncoder()
```

```python
df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
df.head(3)
```

```python
x = df.drop(columns = ['Purchased'])
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 40 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(y_test , y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test , y_pred)
```

```python
df = pd.read_csv("covid_toy.csv")
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer 
```

```python
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender'] = lb.fit_transform(df['gender'])
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 40)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(y_test , y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test , y_pred)
```
